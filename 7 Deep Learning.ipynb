```python
h2o.init()
```

# Load data and split into train and test

```python
url = "http://coursera.h2o.ai/cacao.882.csv"
data = h2o.import_file(url)
```

```python
data.head()
```

```python
train, test = data.split_frame([0.8], seed = 123)
```

# Check enum cardinality to decide which predictors to include

```python
train.nlevels()  #enum cardinality
```

```python
train.shape
```

```python
train.structure()
```

```python
y = "Rating"
# Origin is a substitue column of Bean Origin, can remove
# Maker doesn't contain too much info
ignoreFields = ["Maker", "Origin"]
x = [i for i in train.names if i not in ignoreFields]
```

### A base deep learning model with default settings, runs for 37s, Mean Residual Deviance is 0.1671 in train, 0.2199 in cross validation, 0.2432 in test dataset

```python
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
```

```python
m_def = H2ODeepLearningEstimator(nfolds = 10)
```

```python
%time m_def.train(x, y, train)
```

```python
m_def
```

```python
m_def.model_performance(test)
```

### A tuned model: more hidden neurons with early stopping, runs for 1min 46s, Mean Residual Deviance is 0.1696 in train, 0.2076 in cross validation, 0.2189 in test dataset

```python
m_def.model_performance(test)
```

```python
m_def_400x400 = H2ODeepLearningEstimator(stopping_rounds = 5, #Default
                                         stopping_tolerance = 0, #Default
                                         stopping_metric = "deviance",
                                         hidden = [400, 400],
                                         nfolds = 10)
```

```python
%time m_def_400x400.train(x, y, train)
```

```python
m_def_400x400
```

```python
m_def_400x400.model_performance(test)
```
